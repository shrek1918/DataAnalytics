```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data=pd.read_excel("expense3.xlsx")
df=pd.DataFrame(data)
df["Payment Mode"]=df["Payment Mode"].replace(np.nan,"Cash")
print(df)
```

```python
plt.bar(df["Payment Mode"],df["Amount"])
```

```python
grouped_by=df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)
```

```python
plt.bar(grouped_by.index,grouped_by.values)
plt.show()
```
