```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
dataset = pd.read_csv("pizzas.csv")
df1=pd.DataFrame(dataset)
df = df1.head(35)

# Group the data by pizza type (or ID) and size, then get the mean price
grouped = df.groupby(["pizza_id", "size"])["price"].mean().unstack()

# Fill missing values with 0 (in case some pizza types don't have all sizes)
grouped = grouped.fillna(0)

# Define the pizza types (IDs) for the x-axis
pizza_types = grouped.index

# Stackplot requires the values to be listed by sizes, so extract price columns for each size
sizes = ['S', 'M', 'L', 'XL', 'XXL']  # Assuming these are the only sizes available
price_by_size = [grouped[size] for size in sizes if size in grouped.columns]

# Plot the stackplot
plt.figure(figsize=(10, 6))
plt.stackplot(pizza_types, *price_by_size, labels=sizes)

# Add labels and title
plt.xlabel('Pizza Type')
plt.ylabel('Mean Price')
plt.title('Mean Pizza Prices by Type and Size')
plt.legend(loc='upper left', title="Sizes")

# Show plot
plt.show()
```
